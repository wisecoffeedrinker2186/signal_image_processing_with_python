```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
frequency = 5
amplitude = 5
t = np.linspace(0,10) 
x = amplitude * np.sin(frequency*2*np.pi*t)   # using formula x = Asin(wt+phi)
x_rounded = np.around(x,decimals = 1)         # convert the signal to discrete time by quantising
```

```python
x1 = amplitude * np.cos(frequency*2*np.pi*t)
x_rounded1 = np.around(x1,decimals = 1)
```

**Sinusoidal Signals**

```python
fig, ax = plt.subplots(figsize = (10,6), nrows=2, ncols=2)      
ax[0][0].plot(t,x)
ax[0][1].stem(t,x_rounded, use_line_collection = True)
ax[1][0].plot(t,x1)
ax[1][1].stem(t,x_rounded1, use_line_collection = True);          # stem function to plot the discrete time signal      
ax[0][0].set(title = "Continuous Time Signal")
ax[0][1].set(title = "Discrete Time Signal");
```

**Ramp Signal**

```python
frequency = 5
amplitude = 5
t = np.linspace(0,10) 
x = amplitude*t       # follows formula y = a + bx as ramp or slope signal
x_rounded = np.around(x,decimals = 2)
# alternate method is 
# r = np.where(t<10,t,0)
```

```python
fig, ax = plt.subplots(figsize = (10,6), nrows=1, ncols=2)
ax[0].plot(x)
ax[1].stem(x_rounded);
ax[0].set(title = "Continuous Time Signal")
ax[1].set(title = "Discrete Time Signal")
```

**Step Signal**

```python
frequency = 5
amplitude = 5
t = np.linspace(0,10)
# t1 = int(input("Step length (time): ")) # step length
t1 = 5
x = []
def unit_step(t,t1,x):
    for time in t:
        if  time > t1:
            x.append(0)
        else:
            x.append(1)
unit_step(t,t1,x)
```

```python
# alternate method is 
u = np.where(t<=t1,1,0)     # function is 1 at time < t1 here set to 5

u_rounded = np.around(u,decimals = 2)
fig, ax = plt.subplots(figsize = (10,6), nrows=1, ncols=2)
ax[0].plot(u)
ax[1].stem(u_rounded, use_line_collection=True);
ax[0].set(title = "Continuous Time Signal")
ax[1].set(title = "Discrete Time Signal")
```

**Impluse Signal**

```python
# impulse denoted by delta
t = np.linspace(0,10,8)
impulse = np.where(t==0,1,0)
impulse_rounded = np.around(impulse,decimals = 2)
fig, ax = plt.subplots(figsize = (10,6), nrows=1, ncols=2)
ax[0].plot(impulse)
ax[1].stem(impulse_rounded, use_line_collection = True );
ax[0].set(title = "Continuous Time Signal")
ax[1].set(title = "Discrete Time Signal");
```

**Exponetial Signal**

```python
t = np.linspace(0,10)
a = np.exp(-1*t)              
b = np.exp(t)
fig, ax = plt.subplots(figsize = (10,6), nrows=2, ncols=2)
ax[0][0].plot(a)
ax[0][1].stem(a, use_line_collection = True)
ax[1][0].plot(b)
ax[1][1].stem(b, use_line_collection = True);
ax[0][0].set(title = "Continuous Time Signal")
ax[0][1].set(title = "Discrete Time Signal")
```

**Time Shifting (Advancing Signal)**

```python
t = np.linspace(0,10,10)

u = np.around(np.where(t<=t1,1,0),decimals = 2)
u_advanced = np.around(np.where(t+1<=t1,1,0),decimals = 2)    # advanced 2 ways 

plt.subplot(1, 2, 1)
plt.stem(t, u, use_line_collection = True)
plt.title('u(n)')
plt.axvline(x=0, c="red", label="x=0")
plt.axhline(y=0, c="red", label="y=0")

plt.subplot(1, 2, 2)
plt.stem(t, u_advanced, use_line_collection = True)
plt.title('u(n+1)')
plt.axvline(x=0, c="red", label="x=0")
plt.axhline(y=0, c="red", label="y=0")

plt.show()
```

**Time Reversal**

```python
t = np.linspace(-10,10,10)
u = np.around(np.where(t<=t1,1,0),decimals = 2)
u_delayed = np.around(np.where(t<=t1,1,0),decimals = 2)

plt.subplot(1, 2, 1)
plt.stem(t, u, use_line_collection = True)
plt.title('u(n)')
plt.axvline(x=0, c="red", label="x=0")
plt.axhline(y=0, c="red", label="y=0")

plt.subplot(1, 2, 2)
plt.stem(-t, u, use_line_collection = True)
plt.title('u(-n)')
plt.axvline(x=0, c="red", label="x=0")
plt.axhline(y=0, c="red", label="y=0")

plt.show()
```

# x(n) = a**n
**taking n = 0 to 10 **

```python
t = np.arange(0,10,1)
x_a1 = np.power(-0.5,t) # a is between -1 and 0
x_a2 = np.power(-2,t) # a is < -1
x_a3 = np.power(2,t) # a is > 1
x_a4 = np.power(0.5,t) # a is between 0 and 1

plt.subplot(2, 2, 1)
plt.stem(t, x_a1, use_line_collection = True)
plt.title('-1 < a < 0 ')
plt.axvline(x=0, c="red", label="x=0")
plt.axhline(y=0, c="red", label="y=0")

plt.subplot(2, 2, 2)
plt.stem(t, x_a2, use_line_collection = True)
plt.title('a < -1 ')
plt.axvline(x=0, c="red", label="x=0")
plt.axhline(y=0, c="red", label="y=0")

plt.subplot(2, 2, 3)
plt.stem(t, x_a3, use_line_collection = True)
plt.xlabel('a > 1 \t\t')
plt.axvline(x=0, c="red", label="x=0")
plt.axhline(y=0, c="red", label="y=0")

plt.subplot(2, 2, 4)
plt.stem(t, x_a4, use_line_collection = True)
plt.xlabel(' 0 < a < 1 \t\t')
plt.axvline(x=0, c="red", label="x=0")
plt.axhline(y=0, c="red", label="y=0")

plt.show()
```
