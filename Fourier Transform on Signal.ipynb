```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft,ifft
```

```python
freq = int(input('Enter frequency of signal : '))
sample_freq = int(input('Enter sampling frequency of signal : '))
```

```python
n = np.arange(0,8,0.1)
equation = np.sin(2*np.pi*(freq/sample_freq)*n)
plt.plot(equation)
plt.show();
```

```python
fft_eq = fft(equation)
fft_eq
```

```python
plt.plot(fft_eq)
plt.show();
```

```python
values_calc = []
x= np.array([1,0,0,1])
N =  len(x)
for n in range(len(x)):
    x_k = 0 
    for k in range(len(x)):
        x_k += x[n] * np.exp(-1*1j*2*np.pi*n*k / N)
    values_calc.append(x_k)
print(values_calc)
```

```python
plt.plot(values_calc)
plt.show();
```

If the sampling frequency is less than the frequency of the signal from the figure plotted we can say that there is an inverse relation between the sampling freq. and signal output
